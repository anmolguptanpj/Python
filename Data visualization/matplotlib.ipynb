```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Matplotlib basics

```python
x = np.linspace(0,5,11)
x
y = x**3
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

## subplots in matplotlib

```python
plt.subplot(2,2,1)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

## object oriented way

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the figure object is created. You can use the figsize and dpi keyword arguments. 

figsize is a tuple of widh and hieght of the figure in inches dpi is the dots per inch (pixel per inch)

```python
fig = plt.figure(figsize=(20,8),dpi= 100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2= fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```
