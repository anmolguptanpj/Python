```python
import seaborn as sns
import pandas as pd 
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'vscode'
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
fig = px.line(tips,y='total_bill')
fig.show()
```

```python
avg_tip = tips.groupby('day')['tip'].mean().reset_index()

fig = px.bar(
    avg_tip,
    x='day',
    y='tip',
    title='Average Tip by Day'
)

fig.show()
```

```python
fig = px.scatter(
    tips,
    x='total_bill',
    y='tip'
)

fig.show()
```

```python
fig = px.box(
    tips,
    x='total_bill',
    y='day'
)

fig.show()
```
