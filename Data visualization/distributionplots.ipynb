We use distribution plots to visualize the distribution of quantitative data. 

lets discuss some plots that allow us to visualize the distribution of data set. These plots are: 

* distplot
* jointplot
* pairplot
* rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =  sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

## distplot

```python
sns.displot(df["total_bill"],bins = 20,kde = True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df["tip"],bins = 20,kde = True)
```

## jointplot

```python
sns.jointplot(x='total_bill',y='tip',data = df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data = df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data = df,kind='reg')
```

## pairplot

```python
sns.pairplot(df,hue='size',palette="rainbow")
```

## rugplot

```python
sns.rugplot(df["total_bill"])
```
