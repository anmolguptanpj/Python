## Pandas 

A series is one-dimensional labelled array capable of holding any data type. 
The axis labels are collectively called the index. y

## Creating Series 

There are multiple ways to create a Pandas Series:

```python
import numpy as np 
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d={100:10,200:20,300:30}

```

```python
pd.Series(my_list)
```

```python
 pd.Series(my_list,index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
